```python
import pandas as pd
```

```python
df = pd.read_csv('ML_Project_SNR_Minus5.csv')
```

```python
df
```

```python
df.columns = ['dev_id', 'power', 'sum_real', 'sum_imaginary', 'var_real', 'var_imaginary', 'range_real', 'range_imag', 'real_skew', 'imag_skew', 'real_kurt', 'imag_kurt']
```

```python
df
```

```python
seriesObj = df.apply(lambda x: True if x['dev_id'] > 0 else False , axis=1)
```

```python
numOfRows = len(seriesObj[seriesObj == True].index)
```

```python
print('Number of Rows in dataframe in which signal is transmitted : ', numOfRows)
```

```python
signal_exist = df["dev_id"] > 0
```

```python
df_signal_exist = df.loc[signal_exist]
```

```python
df_signal_exist.shape
```

```python
signal_no_exist = (df["dev_id"] == 0.0)
```

```python
df_signal_no_exist = df.loc[signal_no_exist]
```

```python
df_signal_no_exist.shape
```

```python
df_signal_exist
```

```python
df_signal_no_exist
```

```python
df_signal_exist_5000 = df_signal_exist.head(5000)
```

```python
df_signal_exist_5000.shape
```

```python
df_signal_not_exist_5000 = df_signal_no_exist.head(5000)
```

```python
df_signal_not_exist_5000.shape
```

```python
df_signal_exist_5000.drop(['dev_id'], axis = 1, inplace=True)
```

```python
df_signal_exist_5000.shape
```

```python
df_signal_exist_5000
```

```python
df_signal_not_exist_5000.drop(['dev_id'], axis = 1, inplace=True)
```

```python
df_signal_not_exist_5000.shape
```

```python
df_signal_not_exist_5000
```

```python
df_signal_exist_5000['Signal'] = 1
```

```python
df_signal_exist_5000.shape
```

```python
df_signal_exist_5000
```

```python
df_signal_not_exist_5000['Signal'] = 0
```

```python
df_signal_not_exist_5000.shape
```

```python
df_signal_not_exist_5000
```

```python
concatenate = pd.concat([df_signal_exist_5000, df_signal_not_exist_5000], axis=0)
```

```python
concatenate.shape
```

```python
concatenate
```

```python
concatenate_normalized=(concatenate-concatenate.min())/(concatenate.max()-concatenate.min())
```

```python
concatenate_normalized
```
